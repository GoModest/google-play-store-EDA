# 本notebook的目的是把其中一个软件的评论整理后制作一个词云
## 先来看看哪个软件的评论最多

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import jieba
from wordcloud import WordCloud
from PIL import Image
%matplotlib inline
```

```python
path = r'C:\Users\Ramon\pandas test\data\google-play-store-apps\googleplaystore_user_reviews.csv'
df = pd.read_csv(path)
df.dropna().groupby('App').Translated_Review.size().idxmax()
```

看来评论数最多的Bowmasters,那我们就从这个软件的评论下手来做一个词云

```python
bowmasters = df[df.App == 'Bowmasters'].dropna()
```

```python
res = ' '.join(list(bowmasters.Translated_Review.values))
```

```python
wc = WordCloud(background_color = "white",
               mask = np.array(Image.open(r"image/手柄.jpg")),  #设置背景图片
               max_words = 200, #设置最大显示的字数
               #stopwords = "", #设置停用词
               font_path = r"C:\Windows\Fonts\simhei.ttf",
               max_font_size =200,  #设置字体最大值
               min_font_size = 20,
               random_state = 30, #设置有多少种随机生成状态，即有多少种配色方案
               width = 1000,
               height = 1000
    )
wl = wc.generate(res)
plt.subplots(figsize=(20,20))
plt.imshow(wl)
plt.axis('off')
wl.to_file('image/bowmaster.jpg')
```
